```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.utils import to_categorical

# Daten laden
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalisieren
x_train, x_test = x_train / 255.0, x_test / 255.0

# One-hot-Encoding für Labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Modell erstellen
model = Sequential([
    Flatten(input_shape=(28, 28)),  # Eingabeschicht
    Dense(128, activation='relu'),  # Versteckte Schicht
    Dense(10, activation='softmax')  # Ausgabeschicht
])

# Modell kompilieren
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Modell trainieren
model.fit(x_train, y_train, epochs=5, batch_size=32)

# Modell evaluieren
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"Testgenauigkeit: {test_acc:.4f}")
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import numpy as np

# 1. Daten laden
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 2. Normalisieren
x_train = x_train / 255.0
x_test = x_test / 255.0

# 3. Labels one-hot encodieren
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

# 4. Modell definieren
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# 5. Kompilieren
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# 6. Trainieren mit Validierungsdaten
history = model.fit(x_train, y_train_cat, epochs=5, batch_size=32,
                    validation_split=0.1, verbose=1)

# 7. Evaluieren
test_loss, test_acc = model.evaluate(x_test, y_test_cat)
print(f"Testgenauigkeit: {test_acc:.4f}")

# 8. Beispielvorhersagen visualisieren
predictions = model.predict(x_test[:9])
predicted_labels = np.argmax(predictions, axis=1)

fig, axes = plt.subplots(3, 3, figsize=(9, 9))
for i, ax in enumerate(axes.flat):
    ax.imshow(x_test[i], cmap='gray')
    ax.set_title(f"Echt: {y_test[i]}\nVorhergesagt: {predicted_labels[i]}")
    ax.axis('off')
plt.tight_layout()
plt.show()

# 9. Lernkurven darstellen
fig2, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Genauigkeit
ax1.plot(history.history['accuracy'], label='Training')
ax1.plot(history.history['val_accuracy'], label='Validierung')
ax1.set_title('Genauigkeit pro Epoche')
ax1.set_xlabel('Epoche')
ax1.set_ylabel('Genauigkeit')
ax1.legend()

# Verlust
ax2.plot(history.history['loss'], label='Training')
ax2.plot(history.history['val_loss'], label='Validierung')
ax2.set_title('Verlust pro Epoche')
ax2.set_xlabel('Epoche')
ax2.set_ylabel('Verlust')
ax2.legend()

plt.tight_layout()
plt.show()
```
